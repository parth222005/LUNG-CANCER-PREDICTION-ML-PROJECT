```python
def give_pred(test):
    # Convert the test input to a DataFrame to match the training data format
    columns = ['Age', 'Air Pollution', 'Alcohol use', 'Dust Allergy',
               'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease', # Fixed typo here
               'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker',
               'Chest Pain', 'Coughing of Blood', 'Fatigue', 'Weight Loss',
               'Shortness of Breath', 'Wheezing', 'Swallowing Difficulty',
               'Clubbing of Finger Nails', 'Frequent Cold', 'Dry Cough', 'Snoring',
               'Gender_Female', 'Gender_Male'] # Include one-hot encoded gender columns

    test_df = pd.DataFrame(test, columns=['Age', 'Gender', 'Air Pollution', 'Alcohol use', 'Dust Allergy',
                                          'Occupational Hazards', 'Genetic Risk', 'chronic Lung Disease',
                                          'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker',
                                          'Chest Pain', 'Coughing of Blood', 'Fatigue', 'Weight Loss',
                                          'Shortness of Breath', 'Wheezing', 'Swallowing Difficulty',
                                          'Clubbing of Finger Nails', 'Frequent Cold', 'Dry Cough', 'Snoring'])

    test_df = pd.get_dummies(test_df, columns=['Gender'])

    # Ensure the test data has the same columns as the training data
    test_df = test_df.reindex(columns=columns, fill_value=0)


    prediction_label = ""
    try:
        prediction = classifier.predict(test_df)  # Assuming `classifier` is defined somewhere
        prediction_label = label_encoder.inverse_transform(prediction)[0]
        print("Prediction:", prediction_label)
    except Exception as e:
        print(f"Prediction error: {e}")

    return prediction_label
```
